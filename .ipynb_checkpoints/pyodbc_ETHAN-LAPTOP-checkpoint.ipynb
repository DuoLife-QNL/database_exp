## 使用cursor.description获取执行查询后的列名
> 当前使用的执行方式

```python
def do(sql, query = 1):    
    curs.execute(sql)
    if query == 1:
        row = curs.fetchall()
        column = []
            # row.cursor_description是cursor.description的副本，是一个列表，包含多个tuple，每个tuple对应执行结果的一列，
            # tuple的第一项代表colunm name
            # 但发现（可能是由于pyodbc版本原因）row中并无cursor_description，故直接调用cursor.description
        for t in curs.description:
            column.append(t[0])        #t[0]是每列的列名称
    #         return column
    #     col = get_column()
        d = {}
        for x in column:
            d[x] = ['']
        df = pd.DataFrame(d)
        index = 0
        for x in row:
            df.loc[index, :] = x
            index +=1
        df.index = np.arange(1, len(df) + 1)
        return df
```
